Exercise 3: Rewrite your pay computation with time-and-a-half for overtime and create a function called computepay which takes two parameters (hours and rate).

```python

def computepay(hours, rate):
    if hours > 40:
        regular_pay = 40 * rate
        overtime_pay = (hours - 40) * (rate * 1.5)
        return regular_pay + overtime_pay
    else:
        return hours * rate

try:
  hours = float(input("Enter Hours: "))
  rate = float(input("Enter Rate: "))

  pay = computepay(hours, rate)
  print("Pay:", pay)

except ValueError:
    print("Error, please enter numeric input")
```

Exercise 4: Rewrite the grade program from the previous chapter using a function called computegrade that takes a score as its parameter and returns a grade as a string.

```python
def computegrade(score):
    if score < 0.0 or score > 1.0:
        return "Bad score"
    elif score >= 0.9:
        return "A"
    elif score >= 0.8:
        return "B"
    elif score >= 0.7:
        return "C"
    elif score >= 0.6:
        return "D"
    else:
        return "F"

try:
  score = float(input("Enter Score: "))
  grade = computegrade(score)
  print("Your grade:", grade)

except ValueError:
    print("Bad score")
```

###Part 2: Loops and Iterations

Exercise 1: Write a program which repeatedly reads integers until the user enters “done”. Once “done” is entered, print out the total, count, and average of the integers. If the user enters anything other than a integers, detect their mistake using try and except and print an error message and skip to the next integers.

```python
total = 0
count = 0

while True:
    line = input("> ")

    if line == "done":
        break

    try:
        num = int(line)
        total = total + num
        count = count + 1
    except ValueError:
        print("Invalid input")
        continue

if count > 0:
    average = total / count
else:
    average = 0

print(total,count,average)
```

Exercise 2: Write another program that prompts for a list of numbers as above and at the end prints out both the maximum and minimum of the numbers instead of the average.

```python
largest = None
smallest = None

while True:
    line = input("> ")

    if line == "done":
        break

    try:
        num = int(line)
        if largest is None or num > largest:
            largest = num
        if smallest is None or num < smallest:
            smallest = num

    except ValueError:
        print("Invalid input")
        continue

print("Max: ", largest)
print("Min: ", smallest)
```
